```python
from bs4 import BeautifulSoup as bs
import urllib.request as req
import pandas as pd
```

```python
# 결과를 저장할 리스트
result = []
```

```python
for page in range(1,50):
    Hollys_url = 'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo=%d&sido=&gugun=&store='%page
    print(Hollys_url)
    html = req.urlopen(Hollys_url)
    soupHollys = bs(html, 'html.parser')
    tag_tbody = soupHollys.find('tbody')
    for store in tag_tbody.find_all('tr'):
        if len(store) <=3:
            break
        store_td = store.find_all('td')
        store_sido = store_td[0].string
        store_name = store_td[1].string
        store_name = store_td[1].string
        store_address = store_td[3].string
        store_tel = store_td[5].string
        result.append((store_sido, store_name, store_address, store_tel ))
```

```python
# 크롤링 데이터 DATAFRAME 생성
df = pd.DataFrame(result, columns = ("sido-gu", "store", "address", "tel"))
```

```python
# 크롤링 데이터 저장
df.to_csv("../data/hollys.csv", index=True, encoding="utf-8")
df
```
