```python
# urllib.request 모듈을 읽어 들입니다.
import urllib.request

# urlopen() 함수로 네이버의 메인 페이지를 읽습니다.
res=urllib.request.urlopen('http://www.naver.com')

# 응답된 객체의 타입 출력
print(type(res))

# 응답되면 200 출력 404는 페이지 없음
print(res.status)

# 버전정보 : 11버전
print(res.version)  

# 응답된 객체의 상태 메시지 출력
print(res.msg)

# 응답된 객체의 헤더 정보 출력
print(res.reason)

# 헤더 정보 출력
headers = res.getheaders()
for header in headers:
    print(header)

# 바디 정보 출력
body = res.read()
print(body.decode('utf-8'))
```

```python
# 엘리어스를 활용하여 간단히 작성
import urllib.request as req
res = req.urlopen('http://www.naver.com')
print(res.read().decode('utf-8'))
```

```python
# -- 예외 처리   
# urlerror와 httperror

from urllib.error import URLError, HTTPError
try:
    res = req.urlopen('http://www.naver.com/none') 
except HTTPError as e:
    print(e.code)
    print(e.reason)
    print("HTTPError")
except URLError as e:
    print(e.reason)
    print("URLError")
else:
    print('Success')
```

```python
# parse 모듈을 사용하여 url을 분리
import urllib.parse
url1 = 'http://www.example.com/query?name=김철수&age=30'
url2 = 'http://www.abc.com?name=박문석&age=30'
data1 = urllib.parse.urlparse(url1)
print(data1)
data2 = urllib.parse.urlparse(url2)
print(data2)

# scheme, netloc, path, params, query, fragment에 각각 나누어서 저장됨.
```

-- request 패키지

```python
import requests
url = 'http://www.naver.com'
res = requests.request('get', url)
res.encoding = 'utf-8'
print(res.text)
# 또는
res2 = requests.get(url)
print(res2.text)
```

```python
url1 = 'http://unico2013.dothome.co.kr/crawling/exam.html'

r = requests.get(url1)
r.encoding = 'utf-8'
print(type(r))
if r.text:
    print(r.text)
else:
    print('응답된 콘텐츠가 없어요')
```

```python
url2 = 'http://unico2013.dothome.co.kr/crawling/post.php'

r = requests.post(url2, data ={'name':'백도', 'age':12})
r.encoding = 'utf-8'
print(r.text)
```

```python
# 1. 기상청 중기예보 실습을 requests 라이브러리를 사용하여 경남지역의 일기예보를 출력하라
url = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp'
data = {'stnId':'159'}

params = parse.urlencode(data)
url1 = url + '?' + params
r = requests.get(url1)
r.encoding = 'utf-8'
print(r.text)
```

```python
# 2. 아래의 URL 주소를 사용하여 urllib 라이브러리와 requests 라이브러리를 사용하여 데이터 get 방식으로 로드하여 출력하라.
import urllib.request as req
url = 'https://search.daum.net/nate'
data = {'thr':'sbma', 'w':'tot', 'q':'신지아'}
params = parse.urlencode(data)
url1 = url + '?' + params
res = req.urlopen(url1)
print(res.read().decode('utf-8'))

# 무엇인가 오류가 생기는데 파악 못함. 나중에 다시 확인해보기
```
