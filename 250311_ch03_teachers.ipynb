```python
import bs4
import urllib.request
import csv
import json
import pymysql

aladinUrl = "https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0&cnt=1000&SortOrder=1"
htmlObject = urllib.request.urlopen(aladinUrl)
webPage = htmlObject.read()
bsObject = bs4.BeautifulSoup(webPage, 'html.parser')
#print(bsObject)
```

```python
book_all_100 = bsObject.findAll('div',{'class':'ss_book_list'})
#print(book_all_50)

for index, book in enumerate(book_all_100):
    if index % 2 == 0:
        print(f"index : {index}")
        # print(f"index : {index}\n book{[index]}{book}")
        book_ul = book.find('ul')
        # book_ul_li_list = book_ul.findAll('li')
        book_title = book_ul.find('a', {'class':'bo3'})
        # print(book_ul_li_list)
        print(f"책 제목 : {book_title.text}")

        book_price = ""
        book_author = ""
        book_ul_li_list = book_ul.findAll('li')
        if len(book_ul_li_list) == 5:
            book_price = book_ul_li_list[3].find('span').text
            book_author = book_ul_li_list[2].find('a').text

        elif len(book_ul_li_list) == 4:
            book_price = book_ul_li_list[2].find('span').text
            book_author = book_ul_li_list[1].find('a').text
        print(f"책 가격 : {book_price}")
        print(f"책 저자 : {book_author}")
```

```python
import re
import urllib.request as req
def save_img(title, src):
    pattern = r'[가-힣a-zA-Z0-9]+'  # 한글, 영문 대소문자, 숫자 포함
    result = re.findall(pattern, title)
    result=' '.join(result)
    path=f'images/%s.jpg'%result
    req.urlretrieve(src,path)
```

```python
aladinBase = "https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0&cnt=1000&SortOrder=1&page="

for i in range(10):
    url=aladinBase+str(i+1)
    htmlObject = urllib.request.urlopen(url)
    webPage = htmlObject.read()
    bsObject = bs4.BeautifulSoup(webPage, 'html.parser')
    book_all_100 = bsObject.findAll('div',{'class':'ss_book_list'})
    #print(book_all_50)

    for index, book in enumerate(book_all_100):
        if index % 2 == 0:
            print(f"index : {index*(i+1)}")
            # print(f"index : {index}\n book{[index]}{book}")
            book_ul = book.find('ul')
            # book_ul_li_list = book_ul.findAll('li')
            book_title = book_ul.find('a', {'class':'bo3'})
            # print(book_ul_li_list)
            print(f"책 제목 : {book_title.text}")

            book_price = ""
            book_author = ""
            book_ul_li_list = book_ul.findAll('li')
            if len(book_ul_li_list) == 5:
                book_price = book_ul_li_list[3].find('span').text
                #book_author = book_ul_li_list[2].find('a').text

            elif len(book_ul_li_list) == 4:
                book_price = book_ul_li_list[2].find('span').text
                #book_author = book_ul_li_list[1].find('a').getText()

            book_author = book_ul_li_list[2].find('a')
            if book_author != None:
                book_author=book_author.text
            else:
                book_author=book_ul_li_list[1].find('a').getText()
            
            print(f"책 가격 : {book_price}")
            print(f"책 저자 : {book_author}")
```

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

# 크롬 드라이버 실행
driver = webdriver.Chrome()

# 네이버 접속
driver.get("https://www.naver.com")

# 검색창 찾기
search_box = driver.find_element(By.NAME, "query")
search_box.send_keys("Selenium 자동화", Keys.ENTER)

# "더보기" 버튼 클릭
time.sleep(2)
element1 = driver.find_element(By.CSS_SELECTOR, "a.group_more")
element1.click()

# 검색 결과 가져오기
time.sleep(2)
items = driver.find_elements(By.CSS_SELECTOR, "li.bx")

# 제목 가져오기
for item in items:
    title_element = item.find_element(By.CSS_SELECTOR, "a.link_tit")
    print("제목:", title_element.text)

# 본문 내용 가져오기
for item in items:
    try:
        content_element = item.find_element(By.CSS_SELECTOR, "div.total_dsc_wrap > a")
        print("본문:", content_element.text)
    except:
        print("본문 없음")

# 브라우저 종료
driver.quit()
```
